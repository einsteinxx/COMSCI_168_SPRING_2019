```python
import numpy as np
import matplotlib.pyplot as plt


def compton():
    
    h = 6.62607004e-34 #10-34 #m2 kg / s
    c = 3e8 #speed of light m/s

    
    hv0 = np.arange(1,150,1) #range of incident photon energies
    theta = np.arange(-180,180,1)
    me = 9.10938356e-31 #×10−31 #mass of electron
    joules2ev = 6.242e+18 #1 joule = 6.242e+18 ev 
    
    
    hv_full = np.zeros((hv0.size,theta.size),dtype=float)
    
    for ii in hv0:
        for jj in theta:
            #print(ii,jj)
            hv_cs = hv0[ii-1]/(1 + hv0[ii-1]/(me *(c**2)) * (1 - np.cos(theta[jj-1])))
            hv_full[ii-1,jj-1]=hv_cs
            
     
    
    
    #######
    s = hv_full.shape
    
    for ii in range(0,4): #s[0]): #incident energies
        for jj in range(0,4): #s[1]): #theta output angles
            
            fig = plt.figure()
            fig.suptitle('No axes on this figure')  # Add a title so we know which it is

            fig, ax_lst = plt.subplots(1,1)  # a figure with a 2x2 grid of Axes

            plt.plot(theta[jj],hv_full[ii,jj])
            #plt.plot(e[:,shell-1],label='K shell')
            plt.xlabel('theta angle')
            plt.ylabel('Compton scatter energy')

            plt.legend()
            tname = "theta scatter" + str(theta[jj]) + " Incident energy = " + str(hv0[ii])
            plt.title(tname)
            #plt.show()
            
    return hv_full
```
