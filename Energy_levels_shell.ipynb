```python
# Calculate En for energy shells of atoms
import matplotlib.pyplot as plt
import numpy as np


def kshell():
    Z = np.arange(1,120,1)
    print(Z)
    shells_range = np.arange(1,8,1)
    
    e = np.zeros([Z.size,shells_range.size])
    print(e.size)
    for ii in Z:
        for shell in shells_range:
            e[ii-1,shell-1] = (-1) * -13.6 * (ii**2)/(shell**2)/1000.0 #put into keV
            #print("K = " + str(ii) +  "   shell = " + str(shell))
            
    ### plot results --- so much easier in matlab :(
    
    for ii in Z[0:1]:
        for shell in range(1,8):
            
    
            fig = plt.figure()
            fig.suptitle('No axes on this figure')  # Add a title so we know which it is

            fig, ax_lst = plt.subplots(1,1)  # a figure with a 2x2 grid of Axes

            plt.plot(e[:,shell-1],label='K shell')
            plt.xlabel('Z')
            plt.ylabel('Energy')

            plt.legend()
            tname = "Shell" + str(shell-1) + " Z = " + str(ii)
            plt.title(tname)
            plt.show()
    
    return e
```

```python
e=kshell()
print(e)
```
